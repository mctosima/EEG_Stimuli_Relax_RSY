```python
import numpy
import scipy.signal as signal
import matplotlib.pyplot as plt
```

```python
# create a 8hz sine wave
fs = 100
t = numpy.arange(0, 1, 1/fs)
x = numpy.sin(2*numpy.pi*8*t)

# plot the signal
plt.figure(figsize=(10, 5))
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# create a 13hz sine wave signal as a noise
fs = 100
t = numpy.arange(0, 1, 1/fs)
noise = numpy.sin(2*numpy.pi*40*t)
# reduce the amplitude of the noise signal
# noise = noise * 0.5

# add the noise to the previous signal
x2 = x + noise

# plot the signal
plt.figure(figsize=(10, 5))
plt.plot(t, x2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# create a 13hz sine wave signal as a noise
fs = 100
t = numpy.arange(0, 1, 1/fs)
noise2 = numpy.sin(2*numpy.pi*17*t)
# reduce the amplitude of the noise signal
noise2 = noise2 * 0.5

# add the noise to the previous signal
x2 = x2 + noise2

# plot the signal
plt.figure(figsize=(10, 5))
plt.plot(t, x2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
# do the PSD to the `x2` signal

# get the PSD using FFT method
f, Pxx = signal.welch(x2, fs)

# plot the PSD
plt.figure(figsize=(10, 5))
plt.semilogy(f, Pxx)
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD')
# make y axis range from 0.01 to 1
plt.ylim([0.01, 1])
plt.grid(True)
```
